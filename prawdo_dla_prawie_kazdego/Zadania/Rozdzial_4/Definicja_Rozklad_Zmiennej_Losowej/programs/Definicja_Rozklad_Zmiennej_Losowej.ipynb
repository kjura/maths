```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
import sys
```

### Exercise 4.1.1

```python
Omega = np.array([1, 2, 3, 4, 5, 6])
X = lambda x: np.power(3, x)
range_of_X = np.array(list(map(X, Omega)))

fig, ax = plt.subplots()
ax.stem(Omega, range_of_X, basefmt=' ')
plt.show()
```

### Exercise 4.1.2

```python
Omega = np.array([1, 2, 3, 4, 5, 6])

def X(omega):

    if omega in [3, 5]:
        return -4
    elif omega == 1:
        return 0
    elif omega in [2, 4, 6]:
        return 2
    else:
        sys.exit(f"Aborting, Omega = {Omega} ")

def Prob(X):
    if X == -4:
        return 1 / 3
    elif X == 0:
        return 1 / 6
    elif X == 2:
        return 1 / 2
    else:
        sys.exit(f"Aborting, X takes only -4, 0 or 2 ")


probabilities = np.array(list(map(Prob, [-4, 0, 2])))

fig, ax = plt.subplots()
ax.stem([-4, 0, 2], probabilities, basefmt=' ')
plt.show()
```

### Exercise 4.1.3

```python
omega_count = comb(52, 6)

def probability(X = 0):

    return (comb(13, X) * comb(39, 6 - X)) / omega_count
    
for i in [0, 1, 2, 3, 4, 5, 6]:
    print(f'P(X = {i})', f' = {probability(i):.6f}')

fig, ax = plt.subplots()
ax.stem(np.array([i for i in range(7)]), np.array(list(map(probability, [i for i in range(7)]))), basefmt=' ')
plt.show()
```
