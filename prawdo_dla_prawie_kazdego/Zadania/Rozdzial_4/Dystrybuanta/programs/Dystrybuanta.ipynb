```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
import scipy.integrate as integrate
from matplotlib.patches import Polygon
import sys
```

```python
np.exp(-0.5)
```

```python
def density_expo(x, lmbda):
    return lmbda * np.exp(-1 * lmbda * x)


a, b = 0, 2  # integral limits

lambda_parameter = np.log(4) / 2
non_zero_domain = np.linspace(0, 20, 2000)
non_zero_density = density_expo(non_zero_domain, lmbda=lambda_parameter)

fig, ax = plt.subplots()
ax.plot(np.linspace(-5, 0, 200), np.zeros(200))
ax.plot(non_zero_domain, non_zero_density)

# Make the shaded region
ix = np.linspace(a, b)
iy = density_expo(ix, lambda_parameter)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

ax.text(4 * (a + b), 0.50, r"$\int_{0}^{2} \lambda e^{- \lambda x} dx$",
        horizontalalignment='center', fontsize=20)

ax.spines[['top', 'right']].set_visible(False)
ax.set_xticks([a, b], labels=['$0$', '$2$'])
ax.set_yticks([])

fig.text(0.9, 0.05, '$x$')
fig.text(0.1, 0.9, '$y$')

plt.show()
```
