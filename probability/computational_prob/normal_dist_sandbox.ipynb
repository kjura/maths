$X \sim \mathcal{N}(\mu=3, \sigma=4)$

```python
def pdf_normal(x, mu=0, sigma_squared=1):

    if np.sqrt(sigma_squared) <= 0:
        raise ValueError(f"sigma parameter must be greater than zero but is {sigma_squared}")

    sigma = np.sqrt(sigma_squared)
    scalling_factor = (1 / (sigma * np.sqrt(2 * np.pi)))
    euler_factor = np.power(np.e, (-1 / (2 * np.power(sigma, 2))) * (np.power(x - mu, 2)))

    return scalling_factor * euler_factor


x = np.linspace(-300, 300, 800)
plt.plot(x, pdf_normal(x, mu=4, sigma_squared=100), 'r--')
plt.show()
```
